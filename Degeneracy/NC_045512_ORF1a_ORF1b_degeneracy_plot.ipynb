### R code 

```r
library(ggplot2)
library(zoo)
```

```r
data_ORF1a<-read.csv("ORF1a_degeneracy.txt",sep='\t',header=FALSE)
data_ORF1b<-read.csv("ORF1b_degeneracy.txt",sep='\t',header=FALSE)
```

```r
length(data_ORF1a[,2])
```

```r
length(data_ORF1b[,2])
```

```r
data_ORF1a_coords=data.frame(cbind(1:length(data_ORF1a[,2]),data_ORF1a[,2]))
data_ORF1b_coords=data.frame(cbind(1:length(data_ORF1b[,2]),data_ORF1b[,2]))
```

```r
colnames(data_ORF1a_coords)=c("Position","Degeneracy")
colnames(data_ORF1b_coords)=c("Position","Degeneracy")
```

```r
lab=labs(x="", y="Degeneracy")

theme=theme(plot.title = element_text(size=22),
            axis.title.x = element_text(size = 20),
            axis.title.y = element_text(size = 20),
            axis.text.x = element_text(size = 18), 
            axis.text.y = element_text(size = 18),
            panel.grid.major = element_blank(), 
            panel.grid.minor = element_blank(),
            panel.border = element_rect(size=2),
)
```

```r
p = ggplot(data_ORF1a_coords,aes(Position,Degeneracy))+ 
    geom_line(size=1) +
    lab + labs (title="ORF1a aa degeneracy")+ theme_bw() + 
    xlim(4300, 4406) + 
    theme

print (p)
```

```r
q = ggplot(data_ORF1b_coords,aes(Position,Degeneracy))+ 
    geom_line(size=1) +
    lab + labs (title="ORF1b aa degeneracy")+ theme_bw() + 
    xlim(1, 100) + 
    theme

print (q)
```

```r
window_size=10
```

```r
data_ORF1a_window=rollapply (data_ORF1a[,2], width=10, FUN=mean, align="center")
data_ORF1a_coords_window=data.frame(cbind(1:length(data_ORF1a_window),data_ORF1a_window))

colnames(data_ORF1a_coords_window)=c("Position","Degeneracy")

xlims=xlim(length(data_ORF1a_coords_window[,1])-200, length(data_ORF1a_coords_window[,1]))

p = ggplot(data_ORF1a_coords_window,aes(Position,Degeneracy))+ 
    geom_line(size=1) +
    lab + labs (title=paste("ORF1a aa degeneracy window size",window_size))+ theme_bw() + 
    xlims + 
    theme

print (p)
```

```r
data_ORF1b_window=rollapply (data_ORF1b[,2], width=window_size, FUN=mean, align="center")
data_ORF1b_coords_window=data.frame(cbind(1:length(data_ORF1b_window),data_ORF1b_window))

colnames(data_ORF1b_coords_window)=c("Position","Degeneracy")

p = ggplot(data_ORF1b_coords_window,aes(Position,Degeneracy))+ 
    geom_line(size=1) +
    lab + labs (title=paste("ORF1b aa degeneracy window size",window_size))+ theme_bw() + 
    xlim(0,200)+
    theme

print (p)
```
